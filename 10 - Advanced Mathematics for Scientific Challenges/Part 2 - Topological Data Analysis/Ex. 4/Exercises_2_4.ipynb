# Define and visualize the point cloud $X$ in $\mathbb{R}^{2}$

```python
data = np.array([[1,1], [1,2], [1,3], [2,1], [2,3], [3,1], [3,2], [3,3]])       # Define the point cloud

plt.figure
plt.scatter(data[:,0], data[:,1])             # Plot the point cloud
plt.title("Point cloud $X$")
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.show()
```

# Persistence homology barcode

We can use Gudhi to plot the persisrence homology barcode of the Vietoris-Rips filtration of the point cloud $X$

```python
rips_complex = gudhi.RipsComplex(points=data, max_edge_length=5)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=5)
diag = simplex_tree.persistence(min_persistence=0.1)

gudhi.plot_persistence_barcode(diag)
plt.show()
```

# Persistence diagram

```python
diagrams = ripser(data)['dgms']
plot_diagrams(diagrams, show=True)
```
